```python
# Importing matplotlib library for plotting
import matplotlib.pyplot as plt
```

```python
# Importing numpy to generate data for our plots
import numpy as np
```

```python
# Generating data for sine and cosine waves
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
# Printing the x values
print(x)
```

```python
# Printing the y values (sine of x)
print(y)
```

```python
# Printing the z values (cosine of x)
print(z)
```

Ploting the data

```python
# Plotting the sine wave
plt.plot(x,y)
plt.show()
```

```python
# Plotting the cosine wave
plt.plot(x,z)
plt.show()
```

```python
# Adding title, x-axis, and y-axis labels to the sine wave plot
plt.plot(x,y)
plt.xlabel('angle') # Corrected typo from 'angel' to 'angle'
plt.ylabel('sine value')
plt.title('Sine wave') # Capitalized 'Sine wave' for better readability
plt.show()
```

```python
# Generating data and plotting a parabola
x = np.linspace (-10,10,100)
y = x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola') # Capitalized 'Parabola'
```

```python
# Plotting the parabola with red '+' markers
plt.plot (x , y , 'r+')
plt.show()
```

```python
# Plotting the parabola with green '.' markers
plt.plot(x , y , 'g.')
plt.show()
```

```python
# Generating new data and plotting sine and cosine waves with different styles
x = np.linspace (-5,5,50)
plt.plot(x, np.sin(x), 'g-') # Green solid line for sine
plt.plot(x, np.cos(x), 'r--') # Red dashed line for cosine
plt.show()
```

Bar Plot

```python
# Creating a bar plot of spoken languages and the number of speakers
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100 , 50 ,150 ,40, 70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number of speakers')
plt.title('Spoken Languages')
plt.show()
```

```python
# Creating a pie chart of spoken languages and their distribution
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100 , 50 ,150 ,40, 70]
ax.pie(people, labels = languages, autopct = '%.f%%') # Displaying percentages on the pie chart
plt.show()
```

```python
# Generating data and creating a 2D scatter plot of sine and cosine values
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter (x,y,color='g') # Scattering sine values in green
ax.scatter(x,z,color='b') # Scattering cosine values in blue
plt.show()
```

3D Scatter Plot

```python
# Creating a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d') # Adding a 3D subplot
z= 20 * np.random.random(100) # Generating random z values
x = np.sin(z) # Calculating x values based on sine of z
y = np.cos(z) # Calculating y values based on cosine of z
ax.scatter(x,y,z,c='r',marker='o') # Scattering the points in 3D space with red circles
plt.show()
```

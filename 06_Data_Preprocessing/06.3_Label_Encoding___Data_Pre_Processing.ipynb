Label Encoding:
- converting the labels into numeric form

```python
# importing the Dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
# Loading the data from the CSV file into a pandas DataFrame
cancer_data = pd.read_csv('/content/data.csv')
```

```python
# Display the first 5 rows of the DataFrame
cancer_data.head()
```

```python
# Get the dimensions (rows, columns) of the DataFrame
cancer_data.shape
```

```python
# Find the count of unique values in the 'diagnosis' column
cancer_data['diagnosis'].value_counts()
```

```python
# Initialize the LabelEncoder
label_encode = LabelEncoder()
```

```python
# Apply LabelEncoder to the 'diagnosis' column
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# Append the encoded labels as a new column 'target' to the DataFrame
cancer_data['target'] = labels
```

```python
# Display the first 5 rows of the DataFrame with the new 'target' column
cancer_data.head()
```

```python
# Find the count of unique values in the 'target' column (0= Benign, 1= Malignant)
cancer_data['target'].value_counts()
```

Label Encoding of iris data

```python
# loading the data from csv file to pandas dataFrame
iris_data = pd.read_csv('/content/iris_data.csv')
```

```python
# Display the first 5 rows of the DataFrame
iris_data.head()
```

```python
# Find the count of unique values in the 'Species' column
iris_data['Species'].value_counts()
```

```python
# Initialize the LabelEncoder
label_encoder_1 = LabelEncoder()
```

```python
# Apply LabelEncoder to the 'Species' column and create a new 'target' column
iris_data['target'] = label_encoder_1.fit_transform(iris_data['Species'])
```

```python
# Display the first 5 rows of the DataFrame
iris_data.head()
```

```python
# Display the count of each unique value in the 'target' column
iris_data['target'].value_counts()
```

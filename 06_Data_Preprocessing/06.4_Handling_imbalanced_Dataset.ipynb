**Imbalanced Dataset:**

A dataset with an unequal class distribution

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
```

```python
# Loading the dataset to a pandas DataFrame
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# Displaying the first 5 rows of the DataFrame
credit_card_data.head()
```

```python
# Checking the number of rows and columns in the dataset
credit_card_data.shape
```

```python
# Displaying the last 5 rows of the DataFrame
credit_card_data.tail()
```

```python
# Checking the distribution of the two classes (Legit and Fraudulent Transactions)
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
# Separating the dataset into Legit and Fraudulent Transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
# Printing the shape of the Legit and Fraudulent Transaction DataFrames
print(legit.shape)
print(fraud.shape)
```

Under-sampling

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactionds

Number of Fraudulent Transactions --> 177

```python
# Taking a random sample of 177 Legit Transactions
legit_sample = legit.sample(n=177)
```

```python
# Printing the shape of the Legit Sample DataFrame
print(legit_sample.shape)
```

**Concatenate** the Two DataFrames

```python
# Concatenating the Legit Sample and Fraudulent Transaction DataFrames
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
# Displaying the first 5 rows of the new balanced dataset
new_dataset.head()
```

```python
# Displaying the last 5 rows of the new balanced dataset
new_dataset.tail()
```

```python
# Checking the distribution of the classes in the new balanced dataset
new_dataset['Class'].value_counts()
```

```python
import numpy as np # For numerical operations
import pandas as pd # For data manipulation and analysis
from sklearn.preprocessing import StandardScaler # For standardizing features
from sklearn.model_selection import train_test_split # For splitting data into training and testing sets
from sklearn import svm # For Support Vector Machine algorithm
from sklearn.metrics import accuracy_score # For evaluating model performance
```

Data Collection and Analysis

```python
# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('/content/diabetes (2).csv')
```

```python
# Display the first 5 rows of the dataset
diabetes_dataset.head()
```

```python
# Display the last 5 rows of the dataset
diabetes_dataset.tail()
```

```python
# Get the number of rows and columns in the dataset
diabetes_dataset.shape
```

```python
# Get statistical summary of the dataset , This includes count, mean, std, min, 25%, 50%, 75%, max for numerical columns
diabetes_dataset.describe()
```

```python
# Get the count of each outcome value (0 for Non-Diabetic, 1 for Diabetic)
diabetes_dataset['Outcome'].value_counts()
```

500 --> 0 (Non Diabetic)
268 --> 1 (Diabetic)

```python
# Group the dataset by the 'Outcome' column and calculate the mean of each feature for each group
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Separate features (X) and target (Y)
X = diabetes_dataset.drop(columns = 'Outcome',axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print (X)
print (Y)
```

# Data Standardization



```python
# This scaler will standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
```

```python
# This calculates the mean and standard deviation for each feature
scaler.fit(X)
```

```python
# Transform the features (X) using the fitted scaler, This applies the standardization to the data
standardized_data = scaler.transform(X)
```

```python
# Print the standardized data
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

# Train Test Split

```python
# Split the data into training and testing sets
# test_size=0.2 means 20% of the data will be used for testing
# stratify=Y ensures that the proportion of diabetic and non-diabetic cases is the same in both training and testing sets
# random_state=2 ensures reproducibility of the split
X_train , X_test , Y_train , Y_test = train_test_split(X,Y, test_size= 0.2 , stratify=Y , random_state=2)
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

# Training the Model

```python
# Initialize the Support Vector Machine classifier with a linear kernel
classifier =  svm.SVC(kernel='linear')
```

```python
# Train the Support Vector Machine Classifier
# The model learns from the training data (X_train and Y_train)
classifier.fit(X_train, Y_train)
```

# Model Evaluation

# Accuracy Score

```python
# Make predictions on the training data
X_train_prediction = classifier.predict(X_train)
# Calculate the accuracy score on the training data
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
# Print the accuracy score on the training data
print('Accuracy score on the training data is : ',training_data_accuracy)
```

```python
# Make predictions on the test data
X_test_prediction = classifier.predict(X_test)
# Calculate the accuracy score on the test data
test_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
# Print the accuracy score on the test data
print('Accuracy score on the test data is : ',test_data_accuracy)
```

# Making a Predictive System


```python
# Example of making a prediction on new data

input_data = (1,85,66,29,0,26.6,0.351,31)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
# The input shape needs to be (1, n_features) for prediction
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# standardize the input data
# It's crucial to scale new data using the same scaler fitted on the training data
std_data = scaler.transform(input_data_reshaped)
print('Standardized input data:', std_data)

# prediction
prediction = classifier.predict(std_data)
print('Prediction:', prediction)

# Output the prediction result
if prediction[0] == 0:
    print('The person is not diabetic')
else:
    print('The person is diabetic')
```
